```python
# =================================================================
# PROGRAMA: SISTEMA IA - GESTIÓN DE CIENTÍFICOS
# Estructura: Diccionario principal con lista de registros
# =================================================================

# El diccionario empieza vacío como se solicitó
cientificos = {}

# Inicializamos la llave que contendrá la lista de diccionarios
cientificos["registros"] = []

def mostrar_menu():
    """Muestra las opciones del sistema interactivo."""
    print("\n" + "═"*85)
    print(" 🔬 SISTEMA DE INTELIGENCIA CIENTÍFICA - BASE DE DATOS GLOBAL 🔬 ")
    print("═"*85)
    print(" 1. Registrar nuevo Científico (Entrada de Key-Values)")
    print(" 2. Visualizar Bitácora (Tabla de Investigadores)")
    print(" 3. Ejecutar Diagnóstico de IA (Análisis de Productividad)")
    print(" 4. Finalizar Sesión")
    print("═"*85)

def registrar_cientifico():
    """Función para capturar datos y guardarlos en la lista de diccionarios."""
    print("\n" + "─"*40)
    print("📝 CAPTURA DE NUEVOS DATOS CIENTÍFICOS")
    print("─"*40)

    # Captura de datos (Input de los atributos solicitados)
    rama = input("Rama de la ciencia: ")
    experimentos = input("Cantidad de experimentos activos: ")
    patentes = input("Número de patentes registradas: ")
    presupuesto = input("Presupuesto de investigación ($): ")
    tesis = input("Número de tesis dirigidas: ")

    # --- LISTA DENTRO DEL DICCIONARIO (Equipo de laboratorio) ---
    print("\nIndique 2 equipos esenciales de su laboratorio:")
    equipo_1 = input(" - Equipo principal: ")
    equipo_2 = input(" - Equipo secundario: ")
    lista_equipo = [equipo_1, equipo_2]

    # Crear el diccionario individual con la estructura Key: Value
    datos_cientifico = {
        "Rama": rama,
        "Experimentos": experimentos,
        "Patentes": patentes,
        "Presupuesto": presupuesto,
        "Tesis": tesis,
        "Equipo": lista_equipo  # Atributo que es una lista
    }

    # Agregar el diccionario a la lista que reside dentro del diccionario 'cientificos'
    cientificos["registros"].append(datos_cientifico)
    print("\n✅ DATOS ALMACENADOS CORRECTAMENTE EN EL DICCIONARIO 'cientificos'.")

def imprimir_tabla():
    """Muestra la información en un formato de tabla alineada."""
    lista_activa = cientificos["registros"]

    if not lista_activa:
        print("\n[!] Error: El diccionario de científicos no contiene registros.")
    else:
        # Definición de anchos de columna para la alineación
        print("\n" + "━"*155)
        header = (f"{'RAMA DE CIENCIA':<20} | {'EXP. ACTIVOS':<15} | {'PATENTES':<10} | "
                  f"{'PRESUPUESTO':<15} | {'TESIS DIRIG.':<15} | {'EQUIPO DE LABORATORIO'}")
        print(header)
        print("━"*155)

        # Recorrido de la lista de diccionarios con un ciclo For
        for c in lista_activa:
            # Transformamos la lista de equipo en un solo string estético
            texto_equipo = " / ".join(c['Equipo'])

            fila = (f"{c['Rama']:<20} | {c['Experimentos']:^15} | {c['Patentes']:^10} | "
                    f"${c['Presupuesto']:<14} | {c['Tesis']:^15} | {texto_equipo}")
            print(fila)
        print("━"*155)

def simular_ia():
    """Simulación de IA utilizando condicionales para evaluar a los científicos."""
    lista_activa = cientificos["registros"]

    if not lista_activa:
        print("\n[IA]: No hay datos para procesar el análisis.")
        return

    print("\n[🤖 IA ANALIZANDO PRODUCTIVIDAD CIENTÍFICA...]")
    for c in lista_activa:
        # Usamos condicionales para clasificar según patentes y tesis
        pats = int(c['Patentes'])
        tes = int(c['Tesis'])

        if pats >= 3 and tes >= 10:
            categoria = "Científico Mentor de Élite"
        elif pats > 0:
            categoria = "Investigador con Patentes Activas"
        elif tes > 0:
            categoria = "Investigador Académico"
        else:
            categoria = "Investigador en Etapa Inicial"

        print(f"-> Para la rama de {c['Rama']}, la IA clasifica como: {categoria}.")

# --- BLOQUE DE CONTROL PRINCIPAL (CICLO WHILE TRUE) ---
while True:
    mostrar_menu()
    opcion = input("Seleccione el número de la operación a realizar: ")

    if opcion == "1":
        registrar_cientifico()
    elif opcion == "2":
        imprimir_tabla()
    elif opcion == "3":
        simular_ia()
    elif opcion == "4":
        print("\n[!] Cerrando el Sistema de Gestión Científica. ¡Hasta pronto!")
        break
    else:
        print("\n[!] Opción no reconocida por el sistema. Intente de nuevo.")
```
